```python
import string
from collections import Counter
import matplotlib.pyplot as plt
```

# Before Outbreak

```python
def beforeOutbreak():
    import GetOldTweets3 as got
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('coronavirus covid19') \
        .setSince("2020-02-01") \
        .setUntil("2020-02-29") \
        .setMaxTweets(1000)
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    text_tweets = [[tweet.text] for tweet in tweets]
    return text_tweets
```

```python
before_tweets = beforeOutbreak()
```

```python
text = ""
length = len(before_tweets)
for i in range(0, length):
    text = before_tweets[i][0] + " " + text
lower_case = text.lower()
before_cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))
before_tokenized_words = before_cleaned_text.split()
```

```python
before_tokenized_words
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
before_final_words = [word for word in before_tokenized_words if word not in stop_words]
```

```python
emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        if word in before_final_words:
            emotion_list.append(emotion)

w = Counter(emotion_list)
print(w)
```

```python
fig, ax1 = plt.subplots()
ax1.bar(w.keys(), w.values())
fig.autofmt_xdate()
plt.show()
```
