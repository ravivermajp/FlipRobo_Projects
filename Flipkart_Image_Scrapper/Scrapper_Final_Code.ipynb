```python
from selenium import webdriver
import os
import time
import shutil
import requests
```

```python
driver = webdriver.Chrome(executable_path = 'chromedriver.exe')
my_page = driver.get('https://www.flipkart.com/clothing-and-accessories/bottomwear/trouser/men-trouser/pr?sid=clo%2Cvua%2Cmle%2Clhk&otracker=categorytree&p%5B%5D=facets.occasion%255B%255D%3DFormal&otracker=nmenu_sub_Men_0_Formal%20Trousers')
```

```python
def make_directory(dirname):
    current_path = os.getcwd()
    path = os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)
        
# Image Dir
make_directory("Trowsers")
```

```python
urls = []
x = 0
while(x < 3):
    time.sleep(5)
    images = driver.find_elements_by_xpath("//img[@class='_3togXc']")
    for image in images:
        source = image.get_attribute('src')
        urls.append(source)
        print(source)
    
    button_type=driver.find_element_by_xpath("//div[@class='_2zg3yZ']//a[@class='_3fVaIS']//span").get_attribute('innerHTML')

    if button_type == 'Next':
        driver.find_element_by_xpath("//a[@class='_3fVaIS']").click()
    else:
        driver.find_element_by_xpath("//a[@class='_3fVaIS'][2]").click()

    new_page = driver.find_element_by_xpath("//a[@class='_2Xp0TH fyt9Eu']").text
    
    x=x+1
```

```python
x = 0
while(x < 100):
    print(f'Downloading {x+1} of {100} images')
    #print('Downloading {0} of {1} images'.format(x+1, 100))
    response = requests.get(urls[x])
    with open('Trowsers/trowsers.{0}.jpeg'.format(x+1), "wb") as file:
        file.write(response.content)
    x=x+1
```
